```python
#Connecting MySQL Database
import mysql.connector

host = "127.0.0.1"

user = "root"

password = "Narayan!08"

database = "manufacture"

connection = mysql.connector.connect(

    host=host,

    user=user,

    password=password,

    database=database

)
```

```python
#1 Retrieve ID, name of customers who gave most referrals to new customers
cursor = connection.cursor()
cursor.execute("""SELECT c.customerID, c.customerName, COUNT(r.nr_customerID) AS TotalReferrals
FROM customer c
LEFT JOIN referral r ON c.customerID = r.nr_customerID
GROUP BY c.customerID, c.customerName
ORDER BY TotalReferrals DESC
LIMIT 1""")
cursor.fetchall()
```

```python
#2 Retrieve customers who have placed orders for products manufactured in plants located in New Cody city
cursor = connection.cursor()
cursor.execute("""SELECT DISTINCT c.customerName
FROM customer c
JOIN orders o ON c.customerID = o.o_customerID
JOIN product p ON o.o_plantID = p.p_manufacturingID
JOIN manufacturing m ON p.p_manufacturingID = m.manufacturingID
JOIN plant pl ON m.m_plantID = pl.plantID
WHERE (pl.plantCity = 'New Cody')""")
cursor.fetchall()
```

```python
#3 Retrieve the top 10 products being delivered to customers
cursor = connection.cursor()
cursor.execute("""SELECT p.productName, COUNT(d.d_productID) AS TotalDeliveries
FROM products p
JOIN delivery d ON p.productID = d.d_productID
GROUP BY p.productName
ORDER BY TotalDeliveries DESC
LIMIT 10;""")
cursor.fetchall()
```

```python
#4 Retrieve number of deliveries made to customers
cursor = connection.cursor()
cursor.execute("""SELECT c.customerName, COUNT(d.d_customerID) AS DeliveryCount
FROM customer c
LEFT JOIN delivery d ON c.customerID = d.d_customerID
GROUP BY c.customerName
ORDER BY DeliveryCount DESC;""")
cursor.fetchall()
```

```python
#5 Retrieve the orders made for different product groups
cursor = connection.cursor()
cursor.execute("""SELECT p.productType, COUNT(o.orderID) AS TotalOrders
FROM products p
INNER JOIN orders o ON p.productID = o.o_plantID
GROUP BY p.productType
ORDER BY TotalOrders DESC;""")
cursor.fetchall()
```

```python
#6 Retrieve the customer who have given feedback and also placed orders.
cursor = connection.cursor()
cursor.execute("""SELECT c.customerName
FROM customer c
WHERE EXISTS (
    SELECT 1
    FROM feedback f
    WHERE f.f_customerID = c.customerID
)
AND EXISTS (
    SELECT 1
    FROM orders o
    WHERE o.o_customerID = c.customerID
);""")
cursor.fetchall()
```

```python
#7 Retrieve customer names along with the number of orders they have placed.
cursor = connection.cursor()
cursor.execute("""SELECT c.customerName, COUNT(o.orderID) AS num_orders
FROM customer c
LEFT JOIN orders o ON c.customerID = o.o_customerID
GROUP BY c.customerName;""")
cursor.fetchall()
```

```python
#8 Retrieve customers who have received referrals and the total discount amount they have received.
cursor = connection.cursor()
cursor.execute("""SELECT c.customerName, SUM(r.discountAmount) AS total_discount_amount
FROM customer c
JOIN referral r ON c.customerID = r.nr_customerID
GROUP BY c.customerName;""")
cursor.fetchall()
```

```python
#9 Retrieve product names and types along with the corresponding manufacturing plant:
cursor = connection.cursor()
cursor.execute("""SELECT productName, productType, plantCity
FROM products
JOIN manufacturing ON products.p_manufacturingID = manufacturing.manufacturingID
JOIN plant ON manufacturing.m_plantID = plant.plantID;""")
cursor.fetchall()
```

```python
#10 Retrieve no of orders given to suppliers along with their name and ID
cursor = connection.cursor()
cursor.execute("""SELECT t.tradingID,s.supplierName, COUNT(o.orderID) as Orders
FROM trading t
JOIN orders o ON t.t_plantID = o.o_plantID
JOIN suppliers s ON t.tradingID = s.supplierID
group by t.tradingID,s.supplierName;""")
cursor.fetchall()
```
