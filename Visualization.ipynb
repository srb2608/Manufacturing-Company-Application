```python
import mysql.connector
host = "127.0.0.1"
user = "root"
password = "Narayan!08"
database = "manufacture"
connection = mysql.connector.connect(
 host=host,
 user=user,
 password=password,
 database=database
)
```

```python
cursor = connection.cursor()
cursor.execute("""SELECT orderType, COUNT(*) AS orderTypeCount
FROM orders
GROUP BY orderType;""")
data=cursor.fetchall()
import matplotlib.pyplot as plt
labels, values = zip(*data)

plt.figure(figsize=(8, 8))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Categories')
plt.show()
```

```python
cursor = connection.cursor()
cursor.execute("""SELECT s.supplierName, COUNT(o.orderID) as Orders
FROM trading t
JOIN orders o ON t.t_plantID = o.o_plantID
JOIN suppliers s ON t.tradingID = s.supplierID
group by t.tradingID,s.supplierName;""")
data=cursor.fetchall()
names, values = zip(*data)

plt.figure(figsize=(10, 20))
plt.barh(names, values, color='skyblue')
plt.xlabel('Supplier ID')
plt.title('Supplier ID Distribution')
plt.grid(axis='x')

# Show the plot
plt.show()
```
