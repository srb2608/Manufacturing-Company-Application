```python
import pymongo
import random
# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
database = client["Manufacture"] 

# Create Collections
product_col = database["Products"]
supplier_col = database["Suppliers"]
customer_col = database["Customers"]
delivered_col = database["Delivery"]
```

```python
#Retrieve all products
result = product_col.find({})
for product in result:
    print(product)
```

```python
# Calculate average no of deliveries per customer
result = delivered_col.aggregate([
    {"$group": {"_id": "$D_customer_id", "count": {"$sum": 1}}},
    {"$group": {"_id": None, "average_deliveries": {"$avg": "$count"}}}
])

for avg_deliveries in result:
    print(avg_deliveries)
```

```python
#Customer who have ordered a specefic prducts
product_id = 1  
result = delivered_col.aggregate([
    {"$match": {"D_Product_id": product_id}},
    {"$lookup": {"from": "customer", "localField": "D_customer_id", "foreignField": "CustomerID", "as": "customer_info"}},
    {"$unwind": "$customer_info"},
    {"$project": {"_id": 0, "CustomerInfo": "$customer_info"}}
])

for customer_info in result:
    print(customer_info)
```

```python
#No Of deliveries for each product
result = delivered_col.aggregate([
    {"$group": {"_id": "$D_Product_id", "count": {"$sum": 1}}}
])

for product_count in result:
    print(product_count)
```
